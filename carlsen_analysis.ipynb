```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
y = [2084, 2214, 2385, 2567, 2528, 2675, 2710, 2775, 2772, 2826, 2821, 2837, 2862, 2877, 2853, 2855, 2822, 2842, 2872, 2863]
fig, ax = plt.subplots()
plt.xticks(range(2000, 2021, 2))
ax.ticklabel_format(useOffset=False)
plt.style.use('ggplot')
plt.title("Magnus Carlsen's rating vs Time")
plt.ylabel("Magnus Carlsen's rating")
plt.xlabel("Time (year)")
plt.plot(x, y)
plt.savefig("CARLSENwhole.png")

```
